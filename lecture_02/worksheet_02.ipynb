### Effective Programming

a) What is a drawback of the top down approach?

A drawback of the top down approach is that if there are any changes in the requirement then everything from the top of the tree to the bottom of the tree will be affected and will be directly or indirectly dependent on the change.

b) What is a drawback of the bottom up approach?

A drawback of the bottom up approach is that one may end up working on part of code that may later not be required, hence, it is not the most efficient use of time.

```python
class Board:
  def __init__(self):
    self.board = [["_" for _ in range(8)] for _ in range (8)]
    self.solution_count = 0

  def __repr__(self):
    res = ""
    for row in self.board:
      for col in row:
        res += col
        res += " "
      res += "\n"
    return res

  def is_legal_row(self, row, col):
    for j in range(len(self.board)):
      if self.is_queen(row, j):
        return False
    return True

  def is_legal_col(self, row, col):
    for i in range(len(self.board)):
      if self.is_queen(i, col):
        return False
    return True

  def is_on_board(self, row, col):
    return row >= 0 and row < 8 and col >=0 and col <8

  def is_legal_diag(self, row, col):
    for i in range(len(self.board)):
      if self.is_on_board(row - i, col - i) and self.is_queen(row - i, col - i):
        return False
      if self.is_on_board(row - i, col + i) and self.is_queen(row - i, col + i):
        return False
      if self.is_on_board(row + i, col - i) and self.is_queen(row + i, col - i):
        return False
      if self.is_on_board(row + i, col + i) and self.is_queen(row + i, col + i):
        return False
    return True

  def is_queen(self, row, col):
    if self.board[row][col] == "Q":
      return True
    else:
      return False

  def is_legal(self, row, col):
    if not self.is_legal_row(row, col):
      return False
    if not self.is_legal_col(row, col):
      return False
    if not self.is_legal_diag(row, col):
      return False
    return True

  def set_queen_at(self, row, col):
    self.board[row][col] = "Q"

  def unset_queen_at(self, row, col):
    self.board[row][col] = "_"

  def unset_queen_on(self, row):
    self.board[row] = ["_" for _ in range(8)]

  def get_queen_on(self, row):
    for col in range(len(self.board)):
      if self.is_queen(row, col):
        return col
    raise ValueError("programmer error")


  def search(self, row = 0):

    if row == 8:
      self.solution_count += 1
      print("found a solution")
      print(f"solution number: {self.solution_count}")
      print(self)
      return

    for col in range(8):
      if self.is_legal(row, col):
        self.set_queen_at(row, col)
        self.search(row + 1)
        self.unset_queen_at(row, col)

my_board = Board()
my_board.search()
```
