f) Describe steps of a Data Science Workflow

Mentioned below are the following steps of a ds workflow:
1) Process data: See what data you have, what data is missing, which data to use.
2) Explore data: Understand and visualize the data, find the relation of data with your goal.
3) Extract features: Understand what are the best features possible for the task, do we need new features, do we need to discard some of the features.
4) Create model: Find the correct model, understand the drawbacks and limitations of your model.

This 4 steps repeats, until a satisfactory model is obtained.
